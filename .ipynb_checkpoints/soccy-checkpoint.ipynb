```python
import pandas as pd
import numpy as np

import json

filepath = "data/15946.json"
SHOT_ID = 16

with open(filepath, "r", encoding="UTF-8") as file:
    data = json.load(file)
    
data
```

```python
def getOutcome(event_id):
    event_index = 0
    current_pos = 0
    for i in range(len(data)):
        if data[i].get("id") == event_id:
            
            event_index = i
            current_pos = data[i].get('possession_team').get("id")
            break
            
    for i in range(event_index+1, len(data)):
        event = data[i]
        if event.get("possession_team").get("id") != current_pos:
            return 0
        else if event.get("type").get("id") == SHOT_ID:
            return event.get("shot").get("statsbomb_xg")
    
    return None
getOutcome('1148e967-380f-421f-97f0-d0c820c72984')
    
```
